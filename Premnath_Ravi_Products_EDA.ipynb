```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
products = pd.read_csv('Products.csv')
```

# Display basic information about the dataset

```python
print(products.info())

# Display the first few rows of the dataset
print(products.head())

# Check for missing values
print(products.isnull().sum())

# Summary statistics
print(products.describe())
```

# Analyzing the distribution of product prices

```python
plt.figure(figsize=(10, 6))
sns.histplot(products['Price'], bins=30, kde=True)
plt.title('Price Distribution of Products')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# Analyzing the number of products in each category

```python
category_counts = products['Category'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette='pastel')
plt.title('Number of Products per Category')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.show()
```

# Analyzing the relationship between price and category

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Price', data=products)
plt.title('Price Distribution by Product Category')
plt.xlabel('Category')
plt.ylabel('Price')
plt.show()
```

# Analyzing the top 5 most expensive products

```python
top_5_expensive = products.nlargest(5, 'Price')
plt.figure(figsize=(10, 6))
sns.barplot(x='Price', y='ProductName', data=top_5_expensive, palette='magma')
plt.title('Top 5 Most Expensive Products')
plt.xlabel('Price')
plt.ylabel('Product Name')
plt.show()
```

# Analyzing the top 5 cheapest products

```python
top_5_cheapest = products.nsmallest(5, 'Price')
plt.figure(figsize=(10, 6))
sns.barplot(x='Price', y='ProductName', data=top_5_cheapest, palette='viridis')
plt.title('Top 5 Cheapest Products')
plt.xlabel('Price')
plt.ylabel('Product Name')
plt.show()
```

# Analyzing the average price of products by category

```python
avg_price_by_category = products.groupby('Category')['Price'].mean().sort_values()
plt.figure(figsize=(10, 6))
avg_price_by_category.plot(kind='bar', color='skyblue')
plt.title('Average Price by Product Category')
plt.xlabel('Category')
plt.ylabel('Average Price')
plt.show()
```

# Analyzing the count of products by price range

```python
price_bins = [0, 100, 200, 300, 400, 500, 600]
price_labels = ['0-100', '100-200', '200-300', '300-400', '400-500', '500+']
products['PriceRange'] = pd.cut(products['Price'], bins=price_bins, labels=price_labels)
plt.figure(figsize=(10, 6))
sns.countplot(data=products, x='PriceRange', palette='Set2')
plt.title('Number of Products by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Products')
plt.show()
```
